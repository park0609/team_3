```python
#       
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import squarify
import pandas as pd

d2_path = "./D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
plt.rcParams["font.family"] = "D2Coding"

file_path = "./CARD_SUBWAY_MONTH_202501.csv"
df = pd.read_csv(file_path)
columns = ['name','subway','in','out','day','etx']
df.columns = columns
df = df.loc[:,:"in"]
sdata = df.groupby(by="subway").sum().reset_index()
mdata = sdata[["subway",'in']]
idata = mdata.sort_values(by="in",ascending=[False])
idata.head(10)
plt.figure(figsize=(15,8))
squarify.plot(sizes=idata["in"].iloc[:20],label=idata["subway"].iloc[:20],alpha=0.5)
plt.rcParams['font.size'] = 22
plt.gca().invert_yaxis()
plt.axis("off")
plt.show()
```
