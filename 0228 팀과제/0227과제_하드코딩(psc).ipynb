```python
from collections import Counter
word_counts = Counter(nouns_text.split())
print(word_counts)
```

```python
text = nouns_text
text2 = text.split()
text3 = list(text2)
word = []
count_word = []
result = {}
for n in range(0,len(text3)):
    if not text3[n] in word:
        word.append(text3[n])
        count_word.append(text3.count(text3[n]))
for n in range(0,len(word)):
    result[word[n]] = count_word[n]
print(result)
```
