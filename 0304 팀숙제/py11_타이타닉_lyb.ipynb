```python
# 16 Pclass별 여성 생존자의 수를 출력하시오.
female_survivors = titanic_df[(titanic_df['Sex'] == 'female') & (titanic_df['Survived'] == 1)]
num_female_survivors = len(female_survivors)
Pclass_counts = female_survivors['Pclass'].value_counts()
print(f'타이타닉호 여성 생존자 수: {num_female_survivors}')
for Pclass, count in Pclass_counts.items():
    print(f'{Pclass}등급 여성 생존자: {count}명')
```

```python
# 17 Pclass를 인덱스로 하여 각 컬럼의 개수를 구하시오.
result = titanic_df.groupby('Pclass').count()
result
```
